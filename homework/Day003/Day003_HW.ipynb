```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

import xmltodict

#讀資料
with open('./Data/64_72hr_CH.xml',encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

#找資訊
locations = doc['cwbopendata']['dataset']['locations']['location']

for location in locations:
    print('地區:',location['locationName'])
    print('時間:',location['weatherElement'][0]['time'][0]['dataTime'])
    print('溫度:',location['weatherElement'][0]['time'][0]['elementValue']['value'])
    print('\n')
    

    
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

import xmltodict

#讀資料
with open('./Data/64_72hr_CH.xml',encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))

#找資訊
locations = doc['cwbopendata']['dataset']['locations']['location']

for location in locations:
    print('地區:',location['locationName'])
    loc = doc['cwbopendata']['dataset']['locations']['location']
    
    for w in loc:        
        print('時間:',w['weatherElement'][0]['time'][0]['dataTime'])
        print('溫度:',w['weatherElement'][0]['time'][0]['elementValue']['value'],'\n')
    break
```

# 練習 XML三種套件

```python
# XML.dom
import xml.dom.minidom

doc = xml.dom.minidom.parse('./Data/sample.xml')
print(doc.getElementsByTagName('Title')[0].firstChild.nodeValue)


chapters = doc.getElementsByTagName('Chapter')
for chapter in chapters:
    print(chapter.getAttribute('name'), chapter.firstChild.nodeValue)
```

```python
# XML.etree
import xml.etree.ElementTree as ET

tree = ET.parse('./Data/sample.xml')
root = tree.getroot()

print(root[0].text)

chapters = root[2]
for chapter in chapters:
    print(chapter.attrib['name'],chapter.text)
```

```python
# xmltodict
import xmltodict

with open('./Data/sample.xml',encoding='utf8') as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
# print(doc['CUPOY'])
print(doc['CUPOY']['Title'])

chapters = doc['CUPOY']['Chapters']['Chapter']
# print(chapters)

for chapter in chapters:
    print(chapter['@name'], chapter['#text'])
```
