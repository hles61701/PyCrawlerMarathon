### ① 取出今天所有的發文

```python
import requests
import json
from bs4 import BeautifulSoup as bs 
url = 'https://www.ettoday.net/news/news-list.htm'

req = requests.get(url)
soup = bs(req.text, )

soup
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
for d in (soup.find(class_ = 'part_list_2').find_all('h3')):
    time = d.find(class_ = 'date').text
    tag = d.find(class_ = 'tag').text
    target = d.find(target = '_blank').text
    print(time, '[', tag, ']', target)
```

### ③ 哪一個類別的文章最多 

```python
# 方法一
from collections import Counter
tag_sum = []
for d in (soup.find(class_ = 'part_list_2').find_all('h3')):
    tag = d.find(class_ = 'tag').text
    tag_sum.append(tag)

# 直接使用counters

tag_sum = Counter(tag_sum)
print(tag_sum)
```

```python
# 方法二
tag_sum = []
for d in (soup.find(class_ = 'part_list_2').find_all('h3')):
    tag = d.find(class_ = 'tag').text
    tag_sum.append(tag)
    
tag_dict = {}
for t in tag_sum:
    if t in tag_dict:
        tag_dict[t] +=1
    else : 
        tag_dict[t] = 1
print(tag_dict)

# 將dict的key:value對調=>做降幕排列=>key:value對調回來

[(j, i) for i, j in sorted([(v, k) for k, v in tag_dict.items()], reverse=True)]        
```
