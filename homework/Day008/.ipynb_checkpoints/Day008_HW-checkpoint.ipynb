### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)


r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

type(r.text)
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ',type(soup))

url = 'https://www.dcard.tw/f'
r = requests.get(url)
soup = bs(r.text, 'html')
print(soup.prettify())
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

# 加上headers
h = {'user-agent':'myapp/0.0.1'}

r = requests.get(url, headers = h)

r.encoding = 'utf-8'
print(r.text[0:600])
```
