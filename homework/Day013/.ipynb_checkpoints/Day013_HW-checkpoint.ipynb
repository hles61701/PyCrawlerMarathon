## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.ptt.cc/bbs/NBA/index.html'
req = requests.get(url)
soup = bs(req.text)

soup
```

```python
for d in soup.find_all(class_ ='r-ent'):
    title = d.find(class_ = 'title').find('a').text
    author = d.find(class_ = 'author').text
    date = d.find(class_ = 'date').text
    if str(date) == '11/17':
        print('作者:', author, '標題:', title, '時間:', date )
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup as bs

# page輸入頁數(ex.第一頁:page = 1)
page = 1
url = 'https://www.ptt.cc/bbs/NBA/index{}.html'.format(page)

req = requests.get(url)
soup = bs(req.text)

for d in soup.find_all(class_ ='r-ent'):
    title = d.find(class_ = 'title').find('a').text
    author = d.find(class_ = 'author').text
    date = d.find(class_ = 'date').text
    print('作者:', author, '標題:', title, '時間:', date )
```

```python
# 多頁標題

import requests
from bs4 import BeautifulSoup as bs

# page輸入頁數(ex.第一頁:page = 1)
pages = range(1,10)
for page in pages:
    url = 'https://www.ptt.cc/bbs/NBA/index{}.html'.format(page)

    req = requests.get(url)
    soup = bs(req.text)

    for d in soup.find_all(class_ ='r-ent'):
        title = d.find(class_ = 'title').find('a').text
        author = d.find(class_ = 'author').text
        date = d.find(class_ = 'date').text
        print('作者:', author, '標題:', title, '時間:', date )
```

### ③ 試著爬爬看其他版的文章

```python
import requests
from bs4 import BeautifulSoup as bs


url = 'https://www.ptt.cc/bbs/Gossiping/index.html'

req = requests.get(url, cookies = {'over18' : '1'})
soup = bs(req.text)

for d in soup.find_all(class_ ='r-ent'):
    title = d.find(class_ = 'title').find('a').text
    author = d.find(class_ = 'author').text
    date = d.find(class_ = 'date').text
    print('作者:', author, '標題:', title, '時間:', date )    
```
