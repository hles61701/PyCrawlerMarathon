## 1. 取出今天所有的新聞

```python
# 靜態
import requests
from bs4 import BeautifulSoup as bs

url = 'https://www.ettoday.net/news/news-list.htm'
req = requests.get(url)
soup = bs(req.text)
htmls = soup.find(class_ = 'part_list_2').find_all('h3')
# print(htmls)
for d in htmls:
    print(d.find(class_ = 'date').text, d.find('a').text)
    
```

```python
# 動態
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time

url = 'https://www.ettoday.net/news/news-list.htm'

# 打開瀏覽器
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)

# 每個兩秒鐘自動往下滑,10次
for i in range(15):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
# print(html_source)
soup = bs(html_source)
htmls = soup.find(class_ = 'part_list_2').find_all('h3')
# print(htmls)
for d in htmls:
    print(d.find(class_ = 'date').text, d.find('a').text)

browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import time

timeStamp = time.time()-2*60*60
timeArray = time.localtime(timeStamp)
otherStyleTime = time.strftime("%Y/%m/%d %H:%M", timeArray)
print('現在時間:',otherStyleTime)

url = 'https://www.ettoday.net/news/news-list.htm'

# 打開瀏覽器
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get(url)

# 每個兩秒鐘自動往下滑,10次
for i in range(15):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000);")
    
# 取得資料，丟到 BeautifulSoup 解析
html_source = browser.page_source
# print(html_source)
soup = bs(html_source)
htmls = soup.find(class_ = 'part_list_2').find_all('h3')
# print(htmls)
for d in htmls:
    print(d.find(class_ = 'date').text, d.find('a').text)
    if otherStyleTime >= d.find(class_ = 'date').text:
        break
print('----------stop----------')
browser.quit();
```
